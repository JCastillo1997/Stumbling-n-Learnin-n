Separamos y creamos los .CSV


```python
df = df.rename(columns={
    'famsize': 'n_familiares',
    'sex': 'genero',
    'age': 'edad',
    'address': 'direccion',
    'school': 'escuela',
    'Pstatus': 'estado',
    'Mjob': 'trabajo_m',
    'Fjob': 'trabajo_p',
    'freetime': 'tiempo_libre',
    'Medu': 'educacion_m',
    'Fedu': 'educacion_p',
    'guardian': 'custodia',
    'studytime': 'tiempo_estudio',
    'activities': 'actividades',
    'romantic': 'pareja',
    'goout': 'tiempo_amigos',
    'Dalc': 'alcohol_diario',
    'Walc': 'alcohol_finde',
    'absences': 'faltas'
})
```

```python
df = df.drop(['famsup', 'health', 'nursery', 'higher', 'G1', 'G2', 'G3', 'paid', 'failures', 'schoolsup', 'reason', 'famrel'], axis=1)
```

```python
df = df.drop([ "traveltime","n_familiares"], axis=1)
```

```python
df.info()
```

```python
alumno = df[["genero","edad","direccion","actividades","internet", "pareja","estado","custodia"]]
alumno.direccion = alumno.direccion.str.replace("U","city").str.replace("R","town")
alumno.estado = alumno.estado.str.replace("A","separate").str.replace("T","together")
```

```python
ocio = df[["alcohol_diario","alcohol_finde","tiempo_libre","tiempo_amigos","tiempo_estudio","faltas"]]
```

```python
padres = df[["educacion_m","educacion_p","trabajo_m","trabajo_p"]]
padres
```

```python
colegio = df[["escuela"]]
colegio.escuela = colegio.escuela.str.replace("GP","Gabriel Pereira")
colegio.escuela = colegio.escuela.str.replace("MS","Mousinho da Silveira")
```

```python
alumno.to_csv("alumno.csv")
ocio.to_csv("ocio.csv")
padres.to_csv("padres.csv")
```
